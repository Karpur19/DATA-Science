```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
air_quality = pd.read_csv("C:\\Users\Karpur\OneDrive\Desktop\project\AirqualityUCI.csv", sep=";", decimal=",")
```

```python
# This is our dataset. In this we will specify the parameter for the read function.
```

```python
air_quality.dropna(axis=0, how= 'all', inplace=True)
air_quality.dropna(axis=1, inplace=True)
```

```python
air_quality.replace(to_replace= -200, value= np.NaN, inplace= True)
air_quality.head()
```

```python
# The null values will get replaced in the data.
```

```python
air_quality.fillna(air_quality.mean(), inplace= True)
```

```python
sns.heatmap(air_quality.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
air_quality.loc[:,'Date']=air_quality['Date']
air_quality.head()
```

```python
# We have stored the values in list "Date".
```

```python
from datetime import datetime
Date= []
for i in air_quality['Date']:
    Date.append(datetime.strptime(i,'%d/%m/%Y'))
air_quality.dtypes
```

```python
# Datatypes
```

```python
date=pd.Series(Date)
air_quality['Date']=pd.to_numeric(date)
```

```python
air_quality['Date']=air_quality['Date'].astype(float)
air_quality.dtypes
```

```python
air_quality.loc[:,'Time']=air_quality['Time']
```

```python
# In above we have stored the datatype in list "Time".
```

```python
from datetime import datetime
Time= []
for i in air_quality['Time']:
    Time.append(datetime.strptime(i,'%H.%M.%S'))
air_quality.dtypes
```

```python
time=pd.Series(Time)
air_quality['Time']=pd.to_numeric(time)
```

```python
air_quality['Time']=air_quality['Time'].astype(float)
air_quality.dtypes
```

```python
air_quality.tail()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
num = air_quality.keys()
scaler = MinMaxScaler()
scaler.fit(air_quality[num])
air_quality[num] = scaler.transform(air_quality[num])
```

```python
# MinMaxScaler transforms data in the range between 0 and 1. It is used to convert raw data into a specific range of values for standard use.
```

```python
air_quality['Date']=air_quality['Date'].astype(float)
air_quality.dtypes
air_quality
```

```python
features=air_quality
target=air_quality['T']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=43)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
pred = lin_reg.predict(X_test)
```
